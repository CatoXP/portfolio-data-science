```python
preciosdf = pd.read_csv("Precios Restaurant.csv")
```

```python
preciosdf.head(5)
```

```python
preciosdf["Precio"] = preciosdf["Precio"].str.replace("$","").astype("float") # corregimos str a float
```

```python
preciosdf["Costo"] = preciosdf["Costo"].str.replace("$","").astype("float")# corregimos str a float
```

```python
preciosdf.head(5)
```

```python
preciosdf["Ganancias"] = preciosdf["Precio"]-preciosdf["Costo"]
```

```python
preciosdf.head(5)
```

```python
total_ganancias = sum(preciosdf["Ganancias"])
print(total_ganancias)
```

```python
restdf = restdf.drop("Hora de Cobro", axis=1)
```

```python
restdf
```

```python
precioscarosdf = preciosdf[preciosdf["Precio"]>100]
```

```python
precioscarosdf.head(30)
```

```python
preciosdf["Margen"]=np.where(preciosdf["Ganancias"]>100,"Alto","Bajo")
```

```python
preciosdf.head()
```

```python
restdf = restdf.rename(columns={"Atendió":"Mesero"})
```

```python
restpreciodf = restdf.merge(preciosdf, on=["Producto"],how="left" )
```
