```python
"""

Ejercicio 1: Pila de platos
Crea un programa que simule una pila de platos.
El usuario podrá:
1. Agregar un plato a la pila.
2. Retirar el último plato agregado.
3. Mostrar cuántos platos hay actualmente en la pila.



"""

pila = []

while True:
  print("Menú de pila interactiva")
  print("1 Apilar un elemento push")
  print("2 Desapilar un elemento pop")
  print("3 Mostrar pila")
  print("4 Salir")

  opcion = input("Elige una opción: ")

  if opcion == "1":
    elemento = input("Ingresa el elemento a apilar: ")
    pila.append(elemento)
    print(f"Elemento '{elemento}' apilado correctamente")

  elif opcion == "2":
    if len(pila) == 0:
      print("La pila esta vacía. No hay elementos para desapilar.")
    else:
      retirado = pila.pop()
      print(f"Elemento '{retirado}' retirado del tope")
  elif opcion == "3":
    if len(pila) ==0:
      print("LA pila esta vacia")
    else:
      print("La pila actual (de abajo hacía arriba): ")
      for i, e in enumerate(pila):
        print(f"{i+1}.{e}")

  elif opcion == "4":
    print("Saliendo del programa de pila interactiva")
    break
  else:
    print("Intenta de nuevo")
```

```python
"""
Ejercicio 2: Historial de navegación
Simula el historial de páginas visitadas en un navegador.
Cada vez que el usuario visite una página, agrégala a la pila.
Permite que el usuario “regrese” a la página anterior usando pop().
Muestra siempre la página actual y el historial completo.


"""
pagina = []

while True:
  print(" ----- Bienvenido al historial de navegación ----")
  print("1. ¿Desea visitar la pagina?")
  print("2. ¿Desea regresar a la pagina anterior?")
  print("3. Información (Pagina actual y historial completo)")
  print("4. Salir del historial")

  respuesta = int(input("Ingresa una opción correcta: "))
  if respuesta == 1:
    ingreso = str(input("Ingrese a que pagina va a ingresar: "))
    pagina.append(ingreso)
    print("Su historial a sido registrado correctamente")
  if respuesta == 2:
    if len(pagina) == 0:
      print("El historial esta vacio.")
    else:
      pagina.pop()
      print("Regreso a la pagina anterior")
  if respuesta == 3:
    if len(pagina) == 0:
      print("El historial esta vacio")
    else:
      for a,e in enumerate(pagina):
        print(f"{a+1}# {e} Historial ")
  if respuesta == 4:
    print(" Usted esta saliendo del historial, gracias por visitarnos")
    break
  else:
    print("Intente de nuevo")
```

```python
"""
Ejercicio 3: Deshacer acciones
Crea un programa que permita al usuario escribir acciones (por ejemplo, “dibujar”,
“borrar”, “mover”).
Cada acción se apila.
Permite “deshacer” la última acción usando pop() y mostrar el estado actual de la pila.

"""

dibujo = []

while True:
  print(" ----- Bienvenido al registro de acciones ----")
  print("1. ¿Desea realizar alguna acción")
  print("dibujar,borrar, mover")
  print("2. ¿Desea borrar la ultima accion?")
  print("3. Para ver el histrial de acciones")
  print("4. Salir del historial")

  respuesta = int(input("Ingresa una opción correcta: "))
  if respuesta == 1:
    accion = str(input("Ingresa la acción a realizar: "))
    dibujo.append(accion)
    print(f"Ingresó la acción {accion} del dibujo")

  elif respuesta == 2:
    if len(dibujo) == 0:
      print("No tiene acciones registradas")
    else:
      dibujo.pop()
      print("Ultima acción borrada")
  elif respuesta == 3:
      if len(dibujo) == 0:
        print("No tiene acciones registradas")
      else:
        print("El historial de acciones son estos")
        for a,e in enumerate(dibujo):
          print(f"#{a+1:<3}- Acción realizada: {e:>3} ")

  if respuesta == 4:
    print(" Usted esta saliendo del historial de dibujo.")
    break
  else:
    print("Intente de nuevo")
```

```python

""" Ejercicio 4: Revisión de paréntesis Pide al usuario ingresar una cadena que contenga paréntesis “(” y “)”.
Usa una pila para verificar si los paréntesis están correctamente balanceados.
Ejemplo:  (()) Correcto  (() Incorrecto """


parentesis = []

while True:
    print("\n--- Revisión de paréntesis ---")
    print("1. Ingresar una cadena que contenga paréntesis")
    print("2. Verificar si los paréntesis están correctamente balanceados")
    print("3. Salir del programa")

    try:
        respuesta = int(input("Ingrese la opción deseada: "))
    except ValueError:
        print("Debe ingresar un número válido.")
        continue

    if respuesta == 1:
        cadena = input("Ingrese una cadena que contenga paréntesis '(' y ')' : ")
        parentesis.append(cadena)
        print("✅ La cadena fue ingresada correctamente.")

    elif respuesta == 2:
        if not parentesis:
            print("Primero debe ingresar una cadena.")
        else:
            cadena = parentesis[-1]
            pila = []
            balanceado = True

            for caracter in cadena:
                if caracter == '(':
                    pila.append(caracter)
                elif caracter == ')':
                    if len(pila) == 0:
                        balanceado = False
                        break
                    pila.pop()

            if len(pila) != 0:
                balanceado = False

            if balanceado:
                print(f"'{cadena}' está correctamente balanceada.")
            else:
                print(f" '{cadena}' NO está balanceada.")

    elif respuesta == 3:
        print(" Saliste del programa.")
        break

    else:
        print("Opción no válida, intente de nuevo.")


```

```python
"""
Ejercicio 5: Simulación de elevador
Imagina que cada persona que entra al elevador se apila en la pila.
Cuando alguien sale, se desapila (sale el último que entró).
Permite registrar las entradas y salidas de personas y mostrar el orden actual dentro del
elevador.

"""

elevador = []  # La pila que representa a las personas en el elevador

while True:
    print("\n--- Simulación de elevador ---")
    print("1. Registrar entrada de una persona")
    print("2. Registrar salida de una persona")
    print("3. Mostrar el orden actual en el elevador")
    print("4. Salir del programa")

    try:
        respuesta = int(input("Ingrese la opción deseada: "))
    except ValueError:
        print("Debe ingresar un número válido.")
        continue

    if respuesta == 1:
        persona = input("Ingrese el nombre de la persona que entra al elevador: ")
        elevador.append(persona)
        print(f"{persona} ha entrado al elevador.")

    elif respuesta == 2:
        if len(elevador) == 0:
            print("El elevador está vacío, nadie puede salir.")
        else:
            persona_sale = elevador.pop()
            print(f"{persona_sale} ha salido del elevador.")

    elif respuesta == 3:
        if len(elevador) == 0:
            print("El elevador está vacío.")
        else:
            print("\n Personas en el elevador (de abajo hacia arriba):")
            for i, persona in enumerate(elevador, start=1):
                print(f"{i}. {persona}")

    elif respuesta == 4:
        print(" Saliste del programa.")
        break

    else:
        print(" Opción no válida, intente de nuevo.")

```
