```python
datos = [10, 20, 30, 40, 50, 60, 70]
buscando = 50

# 🔹 Búsqueda Lineal
def busqueda_lineal(lista, valor):
    for i in range(len(lista)):
        if lista[i] == valor:
            return i
    return -1

# 🔹 Búsqueda Binaria
def busqueda_binaria(lista, valor):
    inicio = 0
    fin = len(lista) - 1
    while inicio <= fin:
        medio = (inicio + fin) // 2
        if lista[medio] == valor:
            return medio
        elif lista[medio] < valor:
            inicio = medio + 1
        else:
            fin = medio - 1
    return -1

# 🔹 Resultados
resultado_lineal = busqueda_lineal(datos, buscando)
if resultado_lineal != -1:
    print(f"[Lineal] El valor {buscando} fue encontrado en la posición {resultado_lineal}")
else:
    print(f"[Lineal] El valor {buscando} no está en la lista")

resultado_binaria = busqueda_binaria(datos, buscando)
if resultado_binaria != -1:
    print(f"[Binaria] El valor {buscando} fue encontrado en la posición {resultado_binaria}")
else:
    print(f"[Binaria] El valor {buscando} no está en la lista")
```
