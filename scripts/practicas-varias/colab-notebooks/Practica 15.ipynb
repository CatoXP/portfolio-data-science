```python
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None

class ListaEnlazada:
    def __init__(self):
        self.cabeza = None

    def agregar(self, dato):
        nuevo = Nodo(dato)
        if self.cabeza is None:
            self.cabeza = nuevo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo

    def mostrar(self):
        actual = self.cabeza
        while actual:
            print(actual.dato, end=" --> ")
            actual = actual.siguiente
        print("None")

if __name__ == "__main__":
    lista = ListaEnlazada()


    elementos = ["Daniela", "Javier", "María", "Brenda", "Ivan", "Michell", "Gloria", "David", "Claudia"]
    for nombre in elementos:
        lista.agregar(nombre)

    print("Contenido de la lista enlazada:")
    lista.mostrar()
```
