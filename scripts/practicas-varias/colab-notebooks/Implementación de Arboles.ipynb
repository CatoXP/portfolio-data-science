```python
class Nodo:
    def __init__(self, nombre):
        self.nombre = nombre
        self.hijos = []

    def agregar_hijo(self, nodo_hijo):
        self.hijos.append(nodo_hijo)

    def agregar_hijo_por_nombre(self, nombre_hijo):
        hijo = Nodo(nombre_hijo)
        self.hijos.append(hijo)
        return hijo

    def mostrar(self, nivel=0):
        # Usamos un prefijo visual para simular jerarquía
        print("   " * nivel + f"└─ {self.nombre}")
        for hijo in self.hijos:
            hijo.mostrar(nivel + 1)


if __name__ == "__main__":
    raiz = Nodo("Hogwarts")
    print("-" * 40)

    Gryffindor = raiz.agregar_hijo_por_nombre("Gryffindor")
    Slytherin = raiz.agregar_hijo_por_nombre("Slytherin")
    Ravenclaw = raiz.agregar_hijo_por_nombre("Ravenclaw")
    Hufflepuff = raiz.agregar_hijo_por_nombre("Hufflepuff")

    Gryffindor.agregar_hijo_por_nombre("Harry Potter")
    Gryffindor.agregar_hijo_por_nombre("Hermione Granger")
    Gryffindor.agregar_hijo_por_nombre("Ron Weasley")

    Slytherin.agregar_hijo_por_nombre("Draco Malfoy")
    Slytherin.agregar_hijo_por_nombre("Severus Snape")
    Slytherin.agregar_hijo_por_nombre("Tom Riddle")

    Ravenclaw.agregar_hijo_por_nombre("Luna Lovegood")
    Ravenclaw.agregar_hijo_por_nombre("Cho Chang")

    Hufflepuff.agregar_hijo_por_nombre("Cedric Diggory")
    Hufflepuff.agregar_hijo_por_nombre("Nymphadora Tonks")

    print("Árbol jerárquico de Hogwarts")
    print("-" * 40)
    raiz.mostrar()
```
