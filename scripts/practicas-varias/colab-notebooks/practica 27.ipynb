```python
class Nodo:
    def __init__(self, nombre):
        self.nombre = nombre
        self.hijos = []

    def agregar_hijo(self, hijo):
        self.hijos.append(hijo)

    def mostrar(self, nivel=0):
        # Imprime el nombre con indentación según el nivel
        print("  " * nivel + self.nombre)
        # Llama recursivamente a mostrar() para cada hijo
        for hijo in self.hijos:
            hijo.mostrar(nivel + 1)


# Crear la raíz del árbol
jurassic_park = Nodo("Jurassic Park")

# Crear los grupos de dinosaurios
teropodos = Nodo("Terópodos")
sauropodomorfos = Nodo("Sauropodomorfos")
ornitisquios = Nodo("Ornitisquios")

# Agregar los grupos como hijos de la raíz
jurassic_park.agregar_hijo(teropodos)
jurassic_park.agregar_hijo(sauropodomorfos)
jurassic_park.agregar_hijo(ornitisquios)

# Agregar especies a Terópodos
teropodos.agregar_hijo(Nodo("T-rex"))
teropodos.agregar_hijo(Nodo("Velociraptor"))
teropodos.agregar_hijo(Nodo("Spinosaurus"))

# Agregar especies a Sauropodomorfos
sauropodomorfos.agregar_hijo(Nodo("Brachiosaurus"))
sauropodomorfos.agregar_hijo(Nodo("Diplodocus"))
sauropodomorfos.agregar_hijo(Nodo("Apatosaurus"))

# Agregar especies a Ornitisquios
ornitisquios.agregar_hijo(Nodo("Triceratops"))
ornitisquios.agregar_hijo(Nodo("Stegosaurus"))
ornitisquios.agregar_hijo(Nodo("Iguanodon"))

# Mostrar todo el árbol
jurassic_park.mostrar()
```
