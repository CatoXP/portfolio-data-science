```python
"""
Crear la clase nodo
1-. Cada nodo debe de tener un dato, y un enlace al siguiente nodo
2-. Crear la clase lista enlazada
3-. La lista debe tener un atributo cabeza que inicialmente sea NONE
5-. Implementar un metodo agregar dato que agregue al final de la lista
6-. Implementar un metodo que recorra la lista desde la cabeza
7-. Probar la lista enlazada agregando 3 nombres,

"""
class Nodo:
  def __init__(self,dato):
    self.dato = dato
    self.siguiente = None

class ListaEnlazada:
  def __init__(self):
    self.cabeza = None

  def agregar(self,dato):
    nuevo = Nodo(dato)
    if self.cabeza is None:
      self.cabeza = nuevo
    else:
      actual = self.cabeza
      while actual.siguiente:
        actual = actual.siguiente
      actual.siguiente = nuevo

  def mostrar(self):
    actual = self.cabeza
    while actual:
      print(actual.dato, end="-->")
      actual = actual.siguiente
    print("None")

if __name__ == "__main__":
  lista = ListaEnlazada()

  lista.agregar("Ana")
  lista.agregar("Luis")
  lista.agregar("Marta")

  print("Contenido de la lista: ")
  lista.mostrar()











```
