```python
class Graph:
    def __init__(self):
        self.adj_list = {}

    # Agregar colonia (nodo)
    def agregar_colonia(self, colonia):
        if colonia not in self.adj_list:
            self.adj_list[colonia] = []

    # Agregar ruta (arista)
    def agregar_ruta(self, colonia1, colonia2):
        self.agregar_colonia(colonia1)
        self.agregar_colonia(colonia2)

        self.adj_list[colonia1].append(colonia2)
        self.adj_list[colonia2].append(colonia1)

    # Mostrar grafo
    def mostrar_grafo(self):
        print("\n--- GRAFO COMPLETO ---")
        for colonia, rutas in self.adj_list.items():
            print(f"{colonia}: {rutas}")

    # Ejercicio 2 — verificar conexión
    def esta_conectado(self, x, y):
        if x not in self.adj_list:
            return False
        return y in self.adj_list[x]

    # Ejercicio 3 — grado de una colonia
    def grado(self, colonia):
        if colonia not in self.adj_list:
            return 0
        return len(self.adj_list[colonia])


# ----------------------------------------------------
# Crear grafo inicial
# ----------------------------------------------------
g = Graph()

# Colonias iniciales
g.agregar_ruta("Roma", "Del Valle")
g.agregar_ruta("Roma", "Polanco")
g.agregar_ruta("Del Valle", "Narvarte")
g.agregar_ruta("Narvarte", "Doctores")


# ----------------------------------------------------
# EJERCICIO 1 — Agregar 3 colonias nuevas
# ----------------------------------------------------

g.agregar_ruta("Condesa", "Roma")
g.agregar_ruta("Condesa", "Polanco")

g.agregar_ruta("Coyoacán", "Del Valle")
g.agregar_ruta("Coyoacán", "Narvarte")

g.agregar_ruta("Santa Fe", "Polanco")
g.agregar_ruta("Santa Fe", "Roma")

# Mostrar grafo actualizado
g.mostrar_grafo()


# ----------------------------------------------------
# EJERCICIO 2 — Probar conexiones
# ----------------------------------------------------
print("\n--- PRUEBAS DE CONEXIÓN ---")
print("Roma - Del Valle:", g.esta_conectado("Roma", "Del Valle"))   # True
print("Condesa - Narvarte:", g.esta_conectado("Condesa", "Narvarte")) # False
print("Santa Fe - Polanco:", g.esta_conectado("Santa Fe", "Polanco")) # True


# ----------------------------------------------------
# EJERCICIO 3 — Grado de cada colonia
# ----------------------------------------------------
print("\n--- GRADOS DE CADA COLONIA ---")
mayor = ("", -1)
menor = ("", 9999)

for colonia in g.adj_list:
    grado = g.grado(colonia)
    print(f"{colonia}: {grado} conexiones")

    if grado > mayor[1]:
        mayor = (colonia, grado)

    if grado < menor[1]:
        menor = (colonia, grado)

print("\nColonia con MÁS conexiones:", mayor[0], f"({mayor[1]} rutas)")
print("Colonia con MENOS conexiones:", menor[0], f"({menor[1]} rutas)")
```
