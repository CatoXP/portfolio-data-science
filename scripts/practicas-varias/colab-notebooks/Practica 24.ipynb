```python
# Ejercicio 1: Ordenamiento de números enteros con Quick Sort

# Paso 1: Pedir al usuario una lista de números separados por comas
entrada = input("Ingresa una lista de números enteros separados por comas: ")
numeros = [int(x.strip()) for x in entrada.split(",")]

print("\nLista original:", numeros)

# Paso 2: Implementar Quick Sort
def quick_sort(lista):
    # Caso base: si la lista tiene 1 o 0 elementos, ya está ordenada
    if len(lista) <= 1:
        return lista

    # Seleccionamos el pivote (por simplicidad, el primer elemento)
    pivote = lista[0]

    # Dividimos la lista en dos sublistas:
    # menores: elementos menores al pivote
    # mayores: elementos mayores o iguales al pivote
    menores = [x for x in lista[1:] if x < pivote]
    mayores = [x for x in lista[1:] if x >= pivote]

    # Mostramos el proceso
    print(f"Pivote: {pivote} -> Menores: {menores}, Mayores: {mayores}")

    # Llamamos recursivamente a quick_sort sobre las sublistas
    # La recursión divide el problema en partes más pequeñas
    return quick_sort(menores) + [pivote] + quick_sort(mayores)

# Paso 3: Ordenar la lista y mostrar el resultado
ordenada = quick_sort(numeros)
print("\nLista ordenada (menor a mayor):", ordenada)
```

```python
# Ejercicio 2: Ordenamiento alfabético con Quick Sort

# Puedes definir una lista o pedirla al usuario
# entrada = input("Ingresa palabras separadas por comas: ")
# palabras = [x.strip() for x in entrada.split(",")]
palabras = ["pera", "manzana", "uva", "banana"]

print("\nLista original:", palabras)

def quick_sort_textos(lista):
    if len(lista) <= 1:
        return lista

    pivote = lista[0]
    # En cadenas, Python compara alfabéticamente (según el orden ASCII)
    menores = [x for x in lista[1:] if x.lower() < pivote.lower()]
    mayores = [x for x in lista[1:] if x.lower() >= pivote.lower()]

    print(f"Pivote: {pivote} -> Menores: {menores}, Mayores: {mayores}")

    return quick_sort_textos(menores) + [pivote] + quick_sort_textos(mayores)

ordenadas = quick_sort_textos(palabras)
print("\nLista ordenada alfabéticamente:", ordenadas)
```

```python
# Ejercicio 3: Ordenar productos según su precio con Quick Sort (listas paralelas)

productos = ["Pan", "Leche", "Huevos", "Café"]
precios = [25, 18, 50, 30]

print("Listas originales:")
for p, pr in zip(productos, precios):
    print(f"{p}: ${pr}")

# Quick Sort adaptado para manejar dos listas (productos y precios)
def quick_sort_productos(prod, prec):
    if len(prec) <= 1:
        return prod, prec

    pivote_precio = prec[0]
    pivote_producto = prod[0]

    # Dividimos en dos sublistas: menores y mayores al pivote
    menores_prod = []
    menores_prec = []
    mayores_prod = []
    mayores_prec = []

    # Comparamos los precios y reubicamos productos en sincronía
    for i in range(1, len(prec)):
        if prec[i] < pivote_precio:
            menores_prec.append(prec[i])
            menores_prod.append(prod[i])
        else:
            mayores_prec.append(prec[i])
            mayores_prod.append(prod[i])

    print(f"Pivote: {pivote_producto} (${pivote_precio}) -> Menores: {menores_prec}, Mayores: {mayores_prec}")

    # Llamadas recursivas sobre cada sublista
    menores_prod, menores_prec = quick_sort_productos(menores_prod, menores_prec)
    mayores_prod, mayores_prec = quick_sort_productos(mayores_prod, mayores_prec)

    # Unimos las sublistas en orden correcto
    return (
        menores_prod + [pivote_producto] + mayores_prod,
        menores_prec + [pivote_precio] + mayores_prec
    )

# Ordenamos
productos_ordenados, precios_ordenados = quick_sort_productos(productos, precios)

print("\nListas ordenadas por precio (menor a mayor):")
for p, pr in zip(productos_ordenados, precios_ordenados):
    print(f"{p}: ${pr}")
```
