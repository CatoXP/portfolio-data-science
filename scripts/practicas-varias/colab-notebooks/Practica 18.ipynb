```python

class Nodo:
    def __init__(self, nombre, promedio):
        self.nombre = nombre
        self.promedio = promedio
        self.siguiente = None


class ListaEnlazada:
    def __init__(self):
        self.cabeza = None


    def agregar(self, nombre, promedio):
        nuevo = Nodo(nombre, promedio)
        if self.cabeza is None:
            self.cabeza = nuevo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo


    def agregar_inicio(self, nombre, promedio):
        nuevo = Nodo(nombre, promedio)
        nuevo.siguiente = self.cabeza
        self.cabeza = nuevo


    def eliminar(self, nombre):
        actual = self.cabeza
        anterior = None

        if actual is None:
            return

        if actual.nombre == nombre:
            self.cabeza = actual.siguiente
            return

        while actual and actual.nombre != nombre:
            anterior = actual
            actual = actual.siguiente

        if actual:
            anterior.siguiente = actual.siguiente


    def eliminar_reprobados(self):
        actual = self.cabeza
        anterior = None

        while actual:
            if actual.promedio < 6:
                if anterior:
                    anterior.siguiente = actual.siguiente
                else:
                    self.cabeza = actual.siguiente
            else:
                anterior = actual
            actual = actual.siguiente


    def mostrar(self):
        actual = self.cabeza
        while actual:
            print(f"{actual.nombre} ({actual.promedio})", end=" --> ")
            actual = actual.siguiente
        print("None")



if __name__ == "__main__":
    asistencia = ListaEnlazada()


    asistencia.agregar("daniela", 9.3)
    asistencia.agregar("alfredo", 5.8)
    asistencia.agregar("david", 9.7)
    asistencia.agregar("ximena", 4.3)
    asistencia.agregar("jaime", 6.2)
    asistencia.agregar("rosa", 5.1)
    asistencia.agregar("ivan", 8.6)
    asistencia.agregar("juan", 7.2)
    asistencia.agregar("michell", 5.9)
    asistencia.agregar("andrea", 7.5)
    asistencia.agregar("miguel", 9.8)


    asistencia.agregar_inicio("Profesor Carlos Pérez", 10.0)

    print("Lista de asistencia actualizada:")
    asistencia.mostrar()


    asistencia.eliminar_reprobados()

    print("\n Lista final (sin reprobados):")
    asistencia.mostrar()
```
