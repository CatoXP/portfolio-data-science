```python
def llm(prompt):
    modelo = genai.GenerativeModel("gemini-2.5-flash")
    res = modelo.generate_content(prompt)
    return res.text.strip()
```

```python
print("=== AGENTE IA (Chat + Cotizador por RFC) ===")
print("Escribe algo:\n")

while True:
    user = input("Tú: ").upper().strip()

    # Salir
    if user in ["SALIR", "EXIT", "QUIT"]:
        print("Agente: ¡Hasta luego!")
        break

    # ¿Es un RFC válido en la base?
    if user in EMPRESAS:
        empresa = EMPRESAS[user]

        prompt = f"""
        Genera una cotización profesional para esta empresa:

        Nombre: {empresa['nombre']}
        Giro: {empresa['giro']}
        Descripción: {empresa['descripcion']}

        Incluye:
        - resumen del giro
        - oportunidades comunes del sector
        - propuesta personalizada
        - beneficios claros
        - precio estimado (no final)
        - cierre profesional
        """

        respuesta = llm(prompt)
        print("\n=== COTIZACIÓN ===")
        print(respuesta)
        print("===================\n")

    else:
        # Respuesta normal tipo chat
        prompt = f"""
        Eres un asistente conversacional amable.
        Responde de forma natural a este mensaje:

        Usuario: {user}
        """
        respuesta = llm(prompt)
        print(f"Agente: {respuesta}\n")
```
