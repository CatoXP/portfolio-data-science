```python
"""

 Desarrolla un sistema de atención de clientes de un sistema bancario, donde te atiendan según tu orden de llegada

"""




# Sistema de atención a clientes bancarios (FIFO)

cola_clientes = []

while True:
    print("\n---- Menú de atención a clientes (FIFO) ----")
    print("1. Registrar nuevo cliente")
    print("2. Atender al siguiente cliente")
    print("3. Ver lista de clientes en espera")
    print("4. Salir del sistema")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        nombre = input("Ingresa el nombre del cliente: ")
        motivo = input("Motivo de la visita: ")
        cliente = {"nombre": nombre, "motivo": motivo}
        cola_clientes.append(cliente)
        print(f"Cliente '{nombre}' ha sido registrado en la cola.")

    elif opcion == "2":
        if len(cola_clientes) == 0:
            print("No hay clientes en espera.")
        else:
            cliente_atendido = cola_clientes.pop(0)
            print("\nCliente atendido:")
            print(f"Nombre: {cliente_atendido['nombre']}")
            print(f"Motivo: {cliente_atendido['motivo']}")
            print("Atención finalizada con éxito.")

    elif opcion == "3":
        if len(cola_clientes) == 0:
            print("No hay clientes en cola de espera.")
        else:
            print("\nLista de clientes en espera:")
            for i, cliente in enumerate(cola_clientes, start=1):
                print(f"{i}. {cliente['nombre']} - Motivo: {cliente['motivo']}")

    elif opcion == "4":
        print("Saliendo del sistema de atención al cliente...")
        break

    else:
        print("Opción no válida. Intenta nuevamente.")
```
