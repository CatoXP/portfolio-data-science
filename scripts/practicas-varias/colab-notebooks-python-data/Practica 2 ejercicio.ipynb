```python
# Practica2

"""
Instrucciones
    1.    Crear una clase Persona con atributos nombre y id.
    2.    Crear subclases:
    •    Alumno (con atributo adicional grado)
    •    Profesor (con atributo adicional especialidad)
    3.    Crear clase Libro con atributos titulo y autor.
    4.    Crear clase Prestamo que asocie un Libro con una Persona y tenga un atributo fecha.
    5.    Implementar métodos:
    •    mostrar_info() en Libro y Persona.
    •    realizar_prestamo(libro, persona, fecha) en Prestamo.
    6.    Crear 2 alumnos, 1 profesor, 3 libros y realizar varios préstamos.
    7.    Mostrar la información de cada préstamo.

"""


class Persona:
    def __init__(self, nombre, id):
        self.nombre = nombre
        self.id = id

    def mostrar_info(self):
        print(f"Nombre: {self.nombre}, ID: {self.id}")

class Alumno(Persona):
    def __init__(self, nombre, id, grado):
        super().__init__(nombre, id)
        self.grado = grado

    def mostrar_info(self):
        super().mostrar_info()
        print(f"Grado: {self.grado}")

class Profesor(Persona):
    def __init__(self, nombre, id, especialidad):
        super().__init__(nombre, id)
        self.especialidad = especialidad

    def mostrar_info(self):
        super().mostrar_info()
        print(f"Especialidad: {self.especialidad}")

class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    def mostrar_info(self):
        print(f"Título: {self.titulo}, Autor: {self.autor}")

class Prestamo:
    def __init__(self, libro, persona, fecha):
        self.libro = libro
        self.persona = persona
        self.fecha = fecha

    def mostrar_info(self):
        print("Información del Préstamo:")
        self.libro.mostrar_info()
        self.persona.mostrar_info()
        print(f"Fecha: {self.fecha}")
        print("")


alumno1 = Alumno("juan perez", 101, "10mo grado")

alumno2 = Alumno("maria López", 102, "11vo grado")

profesor1 = Profesor("Dr. Ana Garcia", 201, "Matemáticas")

libro1 = Libro("introducción a python", "guido van rossum")

libro2 = Libro("python avanzado", "mark lutz")

libro3 = Libro("ciencia de datos con python", "wes mcKinney")
```

```python

prestamo1 = Prestamo(libro1, alumno1, "09-22")

prestamo2 = Prestamo(libro2, alumno2, "09-23")

prestamo3 = Prestamo(libro3, profesor1, "09-24")

prestamo4 = Prestamo(libro1, profesor1, "09-25")



prestamo1.mostrar_info()

prestamo2.mostrar_info()


prestamo3.mostrar_info()

prestamo4.mostrar_info()
```
