```python
class Animal:
  def __init__(self,alimentacion,tamaño):
    vivo = True
    self.alimentacion = alimentacion
    self.tamaño = tamaño


class Mamifero(Animal):
  def __init__(self,Piel,Extremidades,Habitad,Tipo_de_sangre,tamaño,alimentacion,vivo):
    super().__init__(alimentacion,tamaño)
    self.piel = Piel
    self.Extremidades = Extremidades
    self.Habitad = Habitad
    self.Tipo_de_sangre = Tipo_de_sangre


class Oviparo(Animal):
  def __init__(self,Huevo,Tipo,Extremidad,alimentacion,tamaño):
    super().__init__(alimentacion,tamaño)
    self.Huevo = Huevo
    self.Tipo = Tipo
    self.Extremidad = Extremidad


class Viviparo(Animal):
  def __init__(self,Genero,Especie,alimentacion,tamaño):
    super().__init__(alimentacion,tamaño)
    self.Genero = Genero
    self.Especie = Especie
```

```python
# Practica 1 LAB

class Persona:
  def __init__(self,nombre,edad):
    self.nombre = nombre
    self.edad = edad

  def saludar(self):
    return f'Hola, soy {self.nombre} y tengo {self.edad} años'


class Profesor(Persona):
  def __init__(self,nombre,edad,especialidad):
    super().__init__(nombre,edad)
    self.especialidad = especialidad

  def enseñar(self):
    return f'{self.nombre} esta enseñando {self.especialidad}'


class Estudiante(Persona):
  def __init__(self,nombre,edad,grado):
    super().__init__(nombre,edad)
    self.grado = grado

  def estudiar(self):
    return f'{self.nombre} esta estudiando {self.grado}'

profesor = Profesor("Ana",40,"Matemáticas")
estudiante = Estudiante("Luis",20,"Segundo Grado")

print(profesor.saludar())
```

```python
# Cursos

class Curso:
  def __init__(self,nombre,alumnos):
    self.nombre = nombre
    self.alumnos = alumnos

  def info(self):
    print(f"Curso: {self.nombre}, alumnos:{self.alumnos}")

class Profesor:
  def __init__(self,nombre,especialidad):
    self.nombre = nombre
    self.especialidad = especialidad
    self.cursos= []

  def asignar_curso(self,curso):
    self.cursos.append(curso)

  def mostrar_cursos(self):
    print(f"Cursos de {self.nombre}:")
    for c in self.cursos:
      print("-",end=" ")
      c.info()

c1 = Curso("Matemáticas",30)
c2 = Curso("Fisica",30)
prof = Profesor("Miguel","Ciencias")
```

```python
prof.asignar_curso(c1)
```

```python
prof.asignar_curso(c2)
```

```python
prof.mostrar_cursos()
```

```python
# Practica2

"""
Instrucciones
    1.    Crear una clase Persona con atributos nombre y id.
    2.    Crear subclases:
    •    Alumno (con atributo adicional grado)
    •    Profesor (con atributo adicional especialidad)
    3.    Crear clase Libro con atributos titulo y autor.
    4.    Crear clase Prestamo que asocie un Libro con una Persona y tenga un atributo fecha.
    5.    Implementar métodos:
    •    mostrar_info() en Libro y Persona.
    •    realizar_prestamo(libro, persona, fecha) en Prestamo.
    6.    Crear 2 alumnos, 1 profesor, 3 libros y realizar varios préstamos.
    7.    Mostrar la información de cada préstamo.

"""


class Persona:
    def __init__(self, nombre, id):
        self.nombre = nombre
        self.id = id

    def mostrar_info(self):
        print(f"Nombre: {self.nombre}, ID: {self.id}")

class Alumno(Persona):
    def __init__(self, nombre, id, grado):
        super().__init__(nombre, id)
        self.grado = grado

    def mostrar_info(self):
        super().mostrar_info()
        print(f"Grado: {self.grado}")

class Profesor(Persona):
    def __init__(self, nombre, id, especialidad):
        super().__init__(nombre, id)
        self.especialidad = especialidad

    def mostrar_info(self):
        super().mostrar_info()
        print(f"Especialidad: {self.especialidad}")

class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    def mostrar_info(self):
        print(f"Título: {self.titulo}, Autor: {self.autor}")

class Prestamo:
    def __init__(self, libro, persona, fecha):
        self.libro = libro
        self.persona = persona
        self.fecha = fecha

    def mostrar_info(self):
        print("Información del Préstamo:")
        self.libro.mostrar_info()
        self.persona.mostrar_info()
        print(f"Fecha: {self.fecha}")
        print("")


alumno1 = Alumno("juan perez", 101, "10mo grado")

alumno2 = Alumno("maria López", 102, "11vo grado")

profesor1 = Profesor("Dr. Ana Garcia", 201, "Matemáticas")

libro1 = Libro("introducción a python", "guido van rossum")

libro2 = Libro("python avanzado", "mark lutz")

libro3 = Libro("ciencia de datos con python", "wes mcKinney")
```

```python


prestamo1 = Prestamo(libro1, alumno1, "2025-09-22")
prestamo2 = Prestamo(libro2, alumno2, "2025-09-23")
prestamo3 = Prestamo(libro3, profesor1, "2025-09-24")
prestamo4 = Prestamo(libro1, profesor1, "2025-09-25")


prestamo1.mostrar_info()
prestamo2.mostrar_info()
prestamo3.mostrar_info()
prestamo4.mostrar_info()
```
