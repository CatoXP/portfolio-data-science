```python
# Primer POO

class persona: # Molde o plantilla en blanco
  def __init__(self,nombre,edad, nacionalidad):
    self.nombre = nombre # atributos
    self.edad = edad  # atributos
    self.nacionalidad = nacionalidad  # atributos

  def saludar(self): # Metodos o acciones
    print(f"Hola me llamo {self.nombre} mi edad es {self.edad} y soy {self.nacionalidad}")

  def cumpleanos(self): # Metodos o acciones
    self.edad +=1
    print(f"Feliz cumpleaños {self.nombre} cumpliste {self.edad} :) ")

ana = persona("Ana",24,"Méxicana") # Creacion del objeto

print(ana.nombre)
print(ana.edad)
print(ana.nacionalidad)
```

```python
ana.cumpleanos() # Feliz cumpleaños Ana cumpliste 25 :)
```

```python
ana.saludar() # Hola me llamo Ana mi edad es 25 y soy Méxicana
```

```python
# segunda POO


class Coche: # plantilla
  def __init__(self,marca,color): # parametros
    self.marca = marca # atributos
    self.color = color
    self.velocidad = 0


  def acelerar(self,cantidad): # metodo acelerar
    self.velocidad += cantidad # acelerar la velocidad
    print(f"El coche {self.marca} va a {self.velocidad} KM/H") # estado actual


  def frenar(self,cantidad): # metodo frenar
    self.velocidad -= cantidad # disminuir velocidad
    if self.velocidad < 0: # Si es menor 0
      self.velocidad = 0 # se frena
    print(f"El coche {self.marca} va a {self.velocidad} KM/H") #estado actual del coche


mi_coche = Coche("toyota","rojo") # creacion del objeto
```

```python
mi_coche.acelerar(50) # estado actual 50
```

```python
mi_coche.acelerar(30) # estado actual
```

```python
mi_coche.frenar(15) # estado actual
```

```python
mi_coche.frenar(100) # frenar por completo
```

```python
# ejemplo 3 POO

class CuentaBancaria:
  def __init__(self,titular,saldo=0):
    self.titular = titular
    self.saldo = saldo

  def depositar(self, cantidad):
    self.saldo += cantidad
    print(f"Su estado de cuenta de {self.titular} es de ${self.saldo}")

  def retirar(self,cantidad):

    if self.saldo < cantidad:
      print("saldo insuficiente")
    else:
      self.saldo -= cantidad
      print(f"Su estado de cuenta de {self.titular} es de ${self.saldo}")

cuenta1 = CuentaBancaria("Juan",100)
```

```python
cuenta1.depositar(5000)
```

```python
cuenta1.retirar(5000)
```
