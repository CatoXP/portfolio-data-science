```python
def display_board(board):
    """
    Muestra el estado actual del tablero en consola.
    """
    print("+-------+-------+-------+")
    for row in board:
        print("|       |       |       |")
        print("|  " + "  |  ".join(str(cell) for cell in row) + "  |")
        print("|       |       |       |")
        print("+-------+-------+-------+")


def enter_move(board):
    """
    Pide al usuario su movimiento, verifica que sea válido (número del 1 al 9 y en casilla libre)
    y actualiza el tablero con una 'O'.
    """
    free = make_list_of_free_fields(board)
    move = None
    while True:
        try:
            move = int(input("Ingresa tu movimiento (1-9): "))
        except ValueError:
            print("Entrada inválida. Ingresa un número del 1 al 9.")
            continue

        # Validar número en rango
        if move < 1 or move > 9:
            print("Número fuera de rango. Intenta de nuevo.")
            continue

        # Buscar en casillas libres
        for row, col in free:
            if board[row][col] == move:
                board[row][col] = 'O'
                return  # termina la función
        print("Esa casilla ya está ocupada. Intenta de nuevo.")


def make_list_of_free_fields(board):
    """
    Recorre el tablero y devuelve una lista de casillas libres.
    Cada casilla libre se guarda como (fila, columna).
    """
    free = []
    for row in range(3):
        for col in range(3):
            if isinstance(board[row][col], int):  # si es número, está libre
                free.append((row, col))
    return free
```
